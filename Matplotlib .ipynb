```python
x = np.arange(10)
print(x)
plt.plot(x,x, label='x vs x', lw = 3, ls = '-', color = "red", marker = '^', markersize = 9, markerfacecolor = 'yellow', markeredgewidth = 3 )
plt.plot(x,x**2, label='x vs x*2',lw = 3, ls = '-', color = "blue", marker = 'o', markersize = 9, markerfacecolor = 'yellow', markeredgewidth = 3)
plt.plot(x,x**3, label='x vs x*3',lw = 3, ls = '-', color = "green", marker = 's', markersize = 9, markerfacecolor = 'yellow', markeredgewidth = 3)
plt.legend(loc = 0) #loc can take values from 0 to 10
plt.show()
```

```python
#Object Oriented method for Plotting.
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.15,0.5,0.4,0.4])
axes1.plot(x,x**2)
axes1.set_title('outer Plot',fontsize=20)
axes1.set_xlabel('xlabel',fontsize=15,rotation=20)
axes1.set_ylabel('ylabel',fontsize=15,rotation=70)

axes2.plot(x**2,x)
axes2.set_title('Inner Plot',fontsize=17)
axes2.set_xlabel('xlabel',fontsize=15)
axes2.set_ylabel('ylabel',fontsize=15)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,x**2)
axes[1].plot(x,x**2)
plt.show()
```
